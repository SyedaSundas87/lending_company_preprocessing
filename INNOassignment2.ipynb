#IMPORT LIBRARIES AND UPLOAD DATASET


```python
import pandas as pd
```

```python
df = pd.read_csv('Lending-company.csv')
df
```

```python
df.tail()
```

```python
df.info()
```

#DROP LOANID AND STRINGID COLUMN

```python
df = df.drop(['LoanID'], axis = 1)
df
```

```python
df = df.drop(['StringID'], axis = 1)
df
```

#Turn the Product categories into numeric data 

```python
df['Product'].unique()
```

```python
mapping = {
    'Product A':0,
    'Product B':1,
    'Product C':2,
    'Product D':2,
    'Product E':2,
    'Product F':2}
df['numeric_Product'] = df['Product'].map(mapping)
df['numeric_Product']
```

```python
df
```

```python
df = df.drop(['Product'], axis = 1)
df
```

#replace â€˜NotSpecifiedâ€™ with the mode of the column

```python
#replace NotSpecified from the mode of column 'CustomerGender'
mode = df['CustomerGender'].mode()[0]
df['Gender'] = df['CustomerGender'].replace('NotSpecified', mode)
```

```python
df
```

```python
df = df.drop(['CustomerGender'], axis = 1)
df
```

#Extract the Location and  Region column numbers and convert them intoe int or float type.

```python
df['Location'] = df['Location'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df['Region'] = df['Region'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
df.head()
```

#HANDLE MISSING VALUES

```python
df.isnull().sum()
```

```python
#fill missing values with mean
df['TotalPrice'] = df['TotalPrice'].fillna(df['TotalPrice'].mean())
df
```

```python
#preprocessing of  â€˜StartDateâ€™ column 

# Drop rows with missing 'StartDate'
df = df.dropna(subset = ['StartDate'])

# Convert 'StartDate' to datetime               
df['StartDate'] = pd.to_datetime(df['StartDate'], format = "%d/%m/%Y")

# Extract month and weekday
df['month'] = df['StartDate'].dt.month
df['weekday'] = df['StartDate'].dt.weekday
df
```

```python
# Drop the 'StartDate' column
df = df.drop(['StartDate'], axis = 1)
df
```

```python
df['TotalDaysYr'].unique
```

```python
df['TotalDaysYr'].nunique()
```

#If a column has the same single value throughout the entire column in a DataFrame, that column does not add meaningful information to the data.No Variability: They do not help in distinguishing data points

```python
df = df.drop(['TotalDaysYr'],axis = 1)
df
```

##preprocessing of LoanStatus 


```python
# Drop rows with missing LoanStatus or where LoanStatus is 'Blocked'
df = df.dropna(subset=['LoanStatus'])
df = df[df['LoanStatus'] != 'Blocked']


# Map 'Active' to 0 and 'Finished Payment' to 1
mapping = {'Active': 0, 'Finished Payment': 1}
df['LoanStatus'] = df['LoanStatus'].map(mapping)
df['LoanStatus']
```
